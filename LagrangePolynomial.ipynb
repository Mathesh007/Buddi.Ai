```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,2,3,5])
y=np.array([2,4,6,10])

plt.plot(x,y,"b-",x,y,"ro")
plt.show()
```

```python
xp=float(input())

yp=0
for xi,yi in zip(x,y):
    yp += yi * np.prod((xp - x[x!=xi])/(xi - x[x!=xi]))
print("Calculated result:",yp)

plt.plot(x,y,"b-",xp,yp,"go",x,y,"ro")
plt.show()
```

```python
from random import randint

x=set()
y=set()

n=0
while n<100:
    xi=randint(-1000,1000)
    if xi not in x:
        n+=1
        x.add(xi)

n=0
while n<100:
    yi=randint(-1000,1000)
    if yi not in y:
        n+=1
        y.add(yi)

x=np.array(list(x),float)
y=list(y)

xplt=np.linspace(x[0],x[-1],num=3000)
yplt=[]

for i in xplt:
    yp=0
    for xi,yi in zip(x,y):
        yp += yi * np.prod((i - x[x!=xi])/(xi - x[x!=xi]))
    yplt.append(yp)

plt.plot(xplt,yplt,"b-")
plt.show()
```
